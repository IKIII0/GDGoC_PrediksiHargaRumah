- sqft_living: luas ruangan yang bisa dihuni. Ini adalah salah satu fitur paling kuat yang berkorelasi positif dengan harga—semakin luas, semakin mahal.

- grade: penilaian kualitas konstruksi rumah. Semakin tinggi nilainya (skala 1-13), semakin bagus kualitas dan semakin mahal rumahnya.

- bathrooms dan bedrooms: jumlah kamar mandi dan kamar tidur juga berpengaruh, walau efeknya tidak sekuat luas bangunan atau grade.

- view dan waterfront: rumah dengan pemandangan menarik atau menghadap air (misal danau) cenderung memiliki harga lebih tinggi.

- zipcode, lat, dan long: lokasi geografis rumah punya peran penting. Beberapa kode pos berkorelasi dengan harga rata-rata yang tinggi karena lingkungan elite atau akses fasilitas.
Lalu ada fitur-fitur yang secara statistik kurang signifikan atau perlu direkayasa ulang:

- yr_built dan yr_renovated: lebih berguna kalau kamu ubah menjadi house_age atau fitur biner seperti is_renovated.

- sqft_lot, sqft_basement, dan sqft_lot15: memiliki korelasi yang rendah dengan harga karena sifatnya tidak langsung mencerminkan nilai rumah.

- condition: walau menunjukan kondisi fisik (skala 1–5), korelasinya dengan harga biasanya lemah dibanding fitur lainnya.

- id dan date: lebih bersifat administratif dan tidak berpengaruh dalam prediksi harga, kecuali jika kamu eksplorasi date lebih dalam untuk analisis musiman.


```python
df = df.drop(columns=['date'])
x = df.drop(columns=['price'])
y = df['price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

##Linear Regression


```python
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
```

```python
y_pred = lr_model.predict(x_test)
r2_lr_score = lr_model.score(x_test, y_test)
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"R²Score: {r2_lr_score:.4f}")
print(f"RMSE: {rmse_lr:.2f}")
```

##Random Forest

```python
rf_model = RandomForestRegressor()
rf_model.fit(x_train, y_train)
```

```python
y_pred = rf_model.predict(x_test)
r2_rf_score = rf_model.score(x_test, y_test)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"R²Score: {r2_rf_score:.4f}")
print(f"RMSE: {rmse_rf:.2f}")
```

- Linear Regression: memberikan interpretasi langsung dan sederhana tentang pengaruh tiap fitur terhadap harga rumah.

- Random Forest: mampu menangkap hubungan non-linier dan interaksi antar fitur dengan stabilitas tinggi.

##Hyperparameter Tuning

Karena disini R²Score dari random forest lebih besar dan RMSE (Root Mean Squared Error) lebih kecil dari model linear regression, maka kita akan men tuning model random forest

```python
rf_model = RandomForestRegressor(n_estimators=200, max_depth=20, random_state=42)
rf_model.fit(x_train, y_train)

rf_score = rf_model.score(x_test, y_test)
print("Akurasi model Random Forest:", rf_score)
```

Setelah di tuning ternyata akurasi model tidak berbeda jauh

```python
# Implementasi menggunakan library pickle
pickle.dump(rf_model, open('Model_PrediksiHargaRumah_RifkiAlSauqy.pkl', 'wb'))
```
